```python
import pandas as pd
import numpy as np
import ta
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('seaborn')

from scipy.stats import linregress
```

```python
df = pd.read_parquet('../Dataframes/df.parquet.gzip')
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.index = df["date"]
df.drop(["date"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split

target = df['delta']
df.drop(['delta'], axis = 1, inplace = True)
data_train, data_test, target_train, target_test = train_test_split(df, target, test_size = 0.30)
```

```python
#import the necessary modules
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#create an object of type SVC
svc_model = SVC(random_state=0)

#train the algorithm on training data and predict using the testing data
pred = svc_model.fit(data_train, target_train).predict(data_test)
estm = svc_model.fit(data_train, target_train)

#print the accuracy score of the model
print("LinearSVC accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
predict_lst = np.array(estm.predict(df.values.tolist()))
predict_lst
```
